```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/train.csv")
df.head()
```

```python
gp = df.groupby("participant_id")
```

```python
participant_ids = []
num_samples = []
```

```python
for participant_id, tmp_df in gp:
    participant_ids.append(participant_id)
    num_samples.append(len(tmp_df))
```

```python
new_df = pd.DataFrame(columns=["participant_ids", "num_samples"])
```

```python
new_df["participant_ids"] = participant_ids
new_df["num_samples"] = num_samples
```

```python
new_df
```

```python
new_df.to_csv("../data/num_samples_per_participants.csv", index=False)
```

```python
participant_label_freq = {}

for participant_id, tmp_df in gp:
    _, label_counts = np.unique(tmp_df.sign.values, axis=None, return_counts=True)
    label_frequencies = label_counts.astype(np.float32) / np.sum(label_counts)
    participant_label_freq[participant_id] = list(label_frequencies)
```
